Something was amis with the PcP-P residuals - we had very large
values for low epicientral distances (< 30 degrees). In this 
notebook I show that this is because I was not handling the 440
and 660 km triplication in P correctly.

We start with the "old" code...

```python
%matplotlib inline
import matplotlib 
import matplotlib.pyplot as plt
import mpl_toolkits.basemap as basemap
# Make the figures big
plt.rcParams['figure.figsize'] = (12, 8)

import pandas
import numpy as np

# Get access to taup
import obspy.taup

# Our ISC parser
import read_ISC

# Read and process the data
all_picks = read_ISC.read_picks('../data/ISC_Jan_2012_trim.dat', ('P', 'PcP'))
paired_picks = read_ISC.pair_picks(all_picks, 'PcP', 'P')
paired_picks_df = pandas.DataFrame.from_dict(paired_picks, orient='index')
paired_picks_df["PcP_ttime"] = paired_picks_df.PcP_datetime - paired_picks_df.event_datetime
paired_picks_df["P_ttime"] = paired_picks_df.P_datetime - paired_picks_df.event_datetime
paired_picks_df["PcP_P_dtime"] = paired_picks_df.PcP_ttime - paired_picks_df.P_ttime

# For plotting timedeltas, we need the delay time in seconds (not as a timedelta64)
paired_picks_df["PcP_P_dtime_secs"] = paired_picks_df.PcP_P_dtime.values / np.timedelta64(1, 's')
```

```python
# Create the Earth model
# iasp91 is built in and 
# distributed with the code
earth_model = obspy.taup.TauPyModel(model="iasp91")

def calc_P_PcP(epi, z):
    arrivals = earth_model.get_travel_times(z, epi, phase_list=['PcP','P'])
    PcP_P_time = arrivals[1].time - arrivals[0].time
    return PcP_P_time

paired_picks_df["PcP_P_dtime_calc_secs"] = paired_picks_df.apply(
    lambda row: calc_P_PcP(row['epicentral_distance'], row['event_depth']), axis=1)

paired_picks_df["PcP_P_resid"] = paired_picks_df.PcP_P_dtime_secs - paired_picks_df.PcP_P_dtime_calc_secs
paired_picks_df.plot(kind='scatter', x='epicentral_distance', y='PcP_P_dtime_calc_secs')
paired_picks_df.plot(kind='scatter', x='epicentral_distance', y='PcP_P_resid')
```

Clearly, the calculated PcP - P time is somehow wrong, giving > 200 sec residuals.

Look at the individual times

```python
def calc_P(epi, z):
    arrivals = earth_model.get_travel_times(z, epi, phase_list=['P'])
    P_ttime = arrivals[0].time
    return P_ttime

def calc_PcP(epi, z):
    arrivals = earth_model.get_travel_times(z, epi, phase_list=['PcP'])
    PcP_ttime = arrivals[0].time
    return PcP_ttime

paired_picks_df["P_ttime_calc_secs"] = paired_picks_df.apply(
    lambda row: calc_P(row['epicentral_distance'], row['event_depth']), axis=1)

paired_picks_df["PcP_ttime_calc_secs"] = paired_picks_df.apply(
    lambda row: calc_PcP(row['epicentral_distance'], row['event_depth']), axis=1)


```

```python
def xmy(x, y):
    return x-y

paired_picks_df["PcP_P_dtime_calc_secs_2"] = paired_picks_df.apply(
    lambda row: xmy(row['PcP_ttime_calc_secs'], row['P_ttime_calc_secs']), axis=1)
```

When handled seperatly, the P and PcP times are fine. 

The problem is that we were getting three P travel times, calling the second "PcP" and thus getting the wrong dtime. 

```python
arrivals = earth_model.get_travel_times(50.0, 24.41, phase_list=['PcP','P']) 
```

```python
print arrivals[0].phase.name
print arrivals[0].time
print arrivals[1].phase.name
print arrivals[1].time
print arrivals[2].phase.name
print arrivals[2].time
print arrivals[3].phase.name
print arrivals[3].time
```

Infact, you can see the triplication. 

```python
P_ttimes = []
eps = []
for epi in np.arange(10,35,0.25):
    arrivals = earth_model.get_travel_times(50.0, epi, phase_list=['P'])
    for arrival in arrivals:
        P_ttimes.append(arrival.time)
        eps.append(epi)
        
plt.plot(eps, P_ttimes, '.')
```

The fix, is to take the first and last (rather than first and second) travel time as P and PcP.
