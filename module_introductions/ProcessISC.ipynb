# Process the ISC data to find PcP-P residual travel times

```python
%matplotlib inline
import matplotlib 
import matplotlib.pyplot as plt
import mpl_toolkits.basemap as basemap
# Make the figures big
plt.rcParams['figure.figsize'] = (12, 8)

import pandas
import numpy as np

# Get access to taup
import obspy.taup

# Our ISC parser
import read_ISC
```

## Read the data

First read the data, and merge picks when an event
has P and PcP at the same station

```python
all_picks = read_ISC.read_picks('../data/ISC_Jan_2012_trim.dat', ('P', 'PcP'))
```

```python
paired_picks = read_ISC.pair_picks(all_picks, 'PcP', 'P')
```

```python
len(paired_picks)
```

## Travel times

Stick the results into a Pandas dataframe and calculate travel times

```python
paired_picks_df = pandas.DataFrame.from_dict(paired_picks, orient='index')
```

```python
paired_picks_df["PcP_ttime"] = paired_picks_df.PcP_datetime - paired_picks_df.event_datetime
paired_picks_df["P_ttime"] = paired_picks_df.P_datetime - paired_picks_df.event_datetime
paired_picks_df["PcP_P_dtime"] = paired_picks_df.PcP_ttime - paired_picks_df.P_ttime

# For plotting timedeltas, we need the delay time in seconds (not as a timedelta64)
paired_picks_df["PcP_P_dtime_secs"] = paired_picks_df.PcP_P_dtime.values / np.timedelta64(1, 's')
```

```python
paired_picks_df
```

```python
paired_picks_df.plot(kind='scatter', x='epicentral_distance', y='PcP_P_dtime_secs')
```

## Model travel times

For each pick pair we need to evaluate the P and PcP time from 
iasp91 and take the difference.

```python
# Create the Earth model
# iasp91 is built in and 
# distributed with the code
earth_model = obspy.taup.TauPyModel(model="iasp91")

def calc_P_PcP(epi, z):
    arrivals = earth_model.get_travel_times(z, epi, phase_list=['PcP','P'])
    # Use the last arrival as PcP - to avoid problems with
    # P triplications.
    PcP_P_time = arrivals[-1].time - arrivals[0].time
    return PcP_P_time

paired_picks_df["PcP_P_dtime_calc_secs"] = paired_picks_df.apply(
    lambda row: calc_P_PcP(row['epicentral_distance'], row['event_depth']), axis=1)
```

Calculate residuals and plot

```python
# -ve is early (w.r.t. model)
paired_picks_df["PcP_P_resid"] = paired_picks_df.PcP_P_dtime_secs - paired_picks_df.PcP_P_dtime_calc_secs
```

```python
paired_picks_df.plot(kind='scatter', x='epicentral_distance', y='PcP_P_resid')
```

```python
plt.hist2d(paired_picks_df.epicentral_distance.values, 
       paired_picks_df.PcP_P_resid.values, 
       bins=40)
plt.colorbar()
plt.show()
```

To put on a map we need the bounce point.

```python
def calc_mid_point_lat(station_lat, station_lon, event_lat, event_lon):
    # see http://www.movable-type.co.uk/scripts/latlong.html
    station_lat = np.radians(station_lat)
    station_lon = np.radians(station_lon)
    event_lat = np.radians(event_lat)
    event_lon = np.radians(event_lon)
    dlon = station_lon-event_lon
    Bx = np.cos(event_lat)*np.cos(dlon)
    By = np.cos(event_lat)*np.sin(dlon)
    mid_lat = np.arctan2(np.sin(station_lat) + np.sin(event_lat),
                    np.sqrt( (np.cos(station_lat)+Bx)*(np.cos(station_lat)+Bx) + By*By ) )
    mid_lat = np.degrees(mid_lat)
    return mid_lat

def calc_mid_point_lon(station_lat, station_lon, event_lat, event_lon):
    # see http://www.movable-type.co.uk/scripts/latlong.html
    station_lat = np.radians(station_lat)
    station_lon = np.radians(station_lon)
    event_lat = np.radians(event_lat)
    event_lon = np.radians(event_lon)
    dlon = station_lon-event_lon
    Bx = np.cos(event_lat)*np.cos(dlon)
    By = np.cos(event_lat)*np.sin(dlon)
    mid_lon = station_lon + np.arctan2(By, np.cos(station_lat) + Bx)
    mid_lon = np.degrees(mid_lon)
    return mid_lon

paired_picks_df["mid_lat"] = paired_picks_df.apply(
    lambda row: calc_mid_point_lat(row['station_lat'], row['station_lon'],
                                   row['event_lat'], row['event_lon'] ), axis=1)

paired_picks_df["mid_lon"] = paired_picks_df.apply(
    lambda row: calc_mid_point_lon(row['station_lat'], row['station_lon'],
                                   row['event_lat'], row['event_lon'] ), axis=1)
```

```python
fig = plt.figure()
axis = fig.add_subplot(1,1,1)
map = basemap.Basemap(projection='hammer',
                      lat_0=0.0, lon_0=180.0, resolution='c', ax=axis)

map.drawcoastlines()
map.drawmapboundary()
map.drawmeridians(np.arange(-180,180,30), labels=[0,0,0,0])
map.drawparallels(np.arange(-90,90,30), labels=[1,0,0,0])

# we need to find the x, y position of the data given
# the map projection.
lonp, latp = map(
    paired_picks_df.mid_lon.values, 
    paired_picks_df.mid_lat.values)

# plot the data
sc = map.scatter(lonp, latp, 
                 c=paired_picks_df.PcP_P_resid.values, 
                 cmap=plt.cm.RdBu, s=50)

plt.colorbar(sc)
plt.show()
```
